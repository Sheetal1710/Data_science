```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df[60]=le.fit_transform(df[60])
```

```python
x = df.iloc[:,0:-1]
x.shape
```

```python
y = df.iloc[:,-1]
y.shape
```

```python
y
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=10)
```

```python
x= pca.fit_transform(x)
```

```python
x.shape
```

```python
pd.DataFrame(data=x)
```

```python
y=y.values.reshape(-1,1)
y.shape
```

```python
from scipy.stats import zscore
z=np.abs(zscore(x))
print(x.shape)
df1=x[(z<3).all(axis=1)]
print(df1.shape)
```

```python
from sklearn.preprocessing import StandardScaler
scale = StandardScaler()
x = scale.fit_transform(x)
```

```python
from sklearn.preprocessing import MinMaxScaler
sc= MinMaxScaler()
x= sc.fit_transform(x)
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix,classification_report
max_accuracy_score=0
for r_state in range(42,100):
    x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=r_state,test_size=0.20)
    lg = LogisticRegression()
    lg.fit(x_train,y_train)
    pred = lg.predict(x_test)
    acc_scr=accuracy_score(y_test,pred)
    print("accuracy_score corresponding to random state:",r_state, " is: ",acc_scr)
    if acc_scr>max_accuracy_score:
        max_accuracy_score=acc_scr
        final_r_state=r_state
        
print("max accuracy score corresponding to",final_r_state," is ",max_accuracy_score)     
```

```python
#Since maximum r2 score is coming around random state 57 we will use that to create our final model
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=57,test_size=0.20)
lg=LogisticRegression()
lg.fit(x_train,y_train)
pred=lg.predict(x_test)
print("accuracy_score:",accuracy_score(y_test,pred))
print (confusion_matrix(y_test,pred))
print (classification_report(y_test,pred))
```

```python
#Checking cross validation
from sklearn.model_selection import cross_val_score
cross_val_score(LogisticRegression(),x,y,cv=10,scoring="accuracy").mean()
```

```python
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
```

```python
auc_score=roc_auc_score(y_test,lg.predict(x_test))
print(auc_score)
```

```python
#For getting best set of parameters of criterion for Dtc model we will use grid search
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier()
grid_param ={
    'criterion': ['gini', 'entropy']
}
gd_sr = GridSearchCV(dtc,grid_param)
gd_sr.fit(x,y)


best_parameters = gd_sr.best_params_
print(best_parameters)
```

```python
#model SVC
from sklearn.svm import SVC
svc = SVC()
grid_param ={'kernel': ('linear','rbf'), 'C':[1,10]}
gd_sr = GridSearchCV(svc,grid_param)
gd_sr.fit(x,y)

best_parameters = gd_sr.best_params_
print(best_parameters)
```

```python
#model: KNeighboursClassifier
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier()
grid_param ={'n_neighbors':range(1,30)}
gd_sr = GridSearchCV(knn,grid_param)
gd_sr.fit(x,y)

best_parameters = gd_sr.best_params_
print(best_parameters)
best_result = gd_sr.best_score_
```

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier()
grid_param={"n_estimators":[10,100,500,1000]}
gd_sr = GridSearchCV(rf,grid_param,cv=5)
gd_sr.fit(x,y)

best_parameters = gd_sr.best_params_
print(best_parameters)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
gb=GradientBoostingClassifier()
grid_param={"learning_rate":[0.001,0.01,0.1,1],"n_estimators":[10,100,500,1000]}
gd_sr = GridSearchCV(gb,grid_param,cv=5)
gd_sr.fit(x,y)

best_parameters = gd_sr.best_params_
print(best_parameters)
```

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.naive_bayes import GaussianNB
gnb=GaussianNB()
ab=AdaBoostClassifier()
grid_param={"learning_rate":[0.001,0.01,0.1,1],"n_estimators":[10,100,500,1000],"base_estimator":[lg,dtc,knn,svc,gnb]}
gd_sr = GridSearchCV(ab,grid_param,cv=5)
gd_sr.fit(x,y)

best_parameters = gd_sr.best_params_
print(best_parameters)
```
