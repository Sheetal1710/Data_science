```python
df_x=df.drop(columns=['Class'])
y=pd.DataFrame(df['Class'])
```

```python
x=df_x
```

```python
from sklearn.preprocessing import StandardScaler
scale = StandardScaler()
x = scale.fit_transform(x)
```

```python
from sklearn.preprocessing import MinMaxScaler
sc= MinMaxScaler()
x= sc.fit_transform(x)
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix,classification_report
max_accuracy_score=0
for r_state in range(42,101):
    x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=r_state,test_size=0.20)
    from sklearn.naive_bayes import MultinomialNB
    mnb=MultinomialNB()
    mnb.fit(x_train,y_train)
    pred = mnb.predict(x_test)
    acc_scr=accuracy_score(y_test,pred)
    print("accuracy_score corresponding to random state:",r_state, " is: ",acc_scr)
    if acc_scr>max_accuracy_score:
        max_accuracy_score=acc_scr
        final_r_state=r_state
        
print("max accuracy score corresponding to",final_r_state," is ",max_accuracy_score)     
```

```python
#Since maximum r2 score is coming around random state 83 we will use that to create our final model
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=83,test_size=0.20)
mnb=MultinomialNB()
mnb.fit(x_train,y_train)
predmnb=mnb.predict(x_test)
print("Predicted result:",pred)
print("actual result",y_test)
print(accuracy_score(y_test,predmnb))
print(confusion_matrix(y_test,predmnb))
print(classification_report(y_test,predmnb))
```

```python
from sklearn.model_selection import cross_val_score
cross_val_score(MultinomialNB(), x,y,cv=5,scoring="accuracy").mean()
```

```python
#For getting best set of parameters we use grid search
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC
svc = SVC()
grid_param ={'kernel': ('linear','poly','rbf'), 'C':[1,10]}
gd_sr = GridSearchCV(svc,grid_param)
gd_sr.fit(x,y)

best_parameters = gd_sr.best_params_
print(best_parameters)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier()
grid_param ={'n_neighbors':range(1,30)}
gd_sr = GridSearchCV(knn,grid_param)
gd_sr.fit(x,y)

best_parameters = gd_sr.best_params_
print(best_parameters)
```

```python
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier()
grid_param ={
    'criterion': ['gini', 'entropy']
}
gd_sr = GridSearchCV(dtc,grid_param)
gd_sr.fit(x,y)

best_parameters = gd_sr.best_params_
print(best_parameters)
```

```python
model=[SVC(kernel='rbf',C=1),KNeighborsClassifier(n_neighbors=7),DecisionTreeClassifier(criterion='gini')]

for m in model:
    m.fit(x_train,y_train)
    m.score(x_train,y_train)
    predm=m.predict(x_test)
    print('Accuracy score of',m,'is:')
    print(accuracy_score(y_test,predm))
    print(confusion_matrix(y_test,predm))
    print(classification_report(y_test,predm))
    print('\n')
```

```python
print("Mean accuracy score for KNeighbors:",cross_val_score(KNeighborsClassifier(), x,y,cv=5,scoring="accuracy").mean())
print("Mean accuracy score for SVC:",cross_val_score(SVC(), x,y,cv=10,scoring="accuracy").mean())
print("Mean accuracy score for DecisionTree:",cross_val_score(DecisionTreeClassifier(), x,y,cv=5,scoring="accuracy").mean())
```
